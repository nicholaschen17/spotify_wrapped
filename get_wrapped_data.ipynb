```python
def get_track_ids(time_frame):
    track_ids = []
    for song in time_frame['items']:
        track_ids.append(song['id'])
    return track_ids
```

```python
def get_artists(time_frame):
    artists = []
    for artist in time_frame['items']:
        artists.append(artist['id'])
    return artists
```

```python
def get_track_features(id):
    meta = sp.track(id)
    # meta
    name = meta['name']
    album = meta['album']['name']
    artist = meta['album']['artists'][0]['name']
    spotify_url = meta['external_urls']['spotify']
    album_cover = meta['album']['images'][0]['url']
    track_info = [name, album, artist, spotify_url, album_cover]
    return track_info
```

```python
def get_artist_features(id):
    meta = sp.artist(id)
    # meta
    artist_name = meta['name']
    artist_type = meta['type']
    artist_genres = meta['genres']
    artist_image = meta['images'][0]['url']
    artist_info = [artist_name, artist_type, artist_genres, artist_image]
    return artist_info
```

```python
def convert_tracks_to_df(track_ids, time_period, today):
    # loop over track ids 
    tracks = []
    for i in range(len(track_ids)):
        time.sleep(.5)
        track = get_track_features(track_ids[i])
        tracks.append(track)
        
    # create dataset
    df = pd.DataFrame(tracks, columns = ['name', 'album', 'artist', 'spotify_url', 'album_cover'])
    # save to CSV
    df.to_json(f'{today}-{time_period} songs.json')
```

```python
time_ranges = ['short_term', 'medium_term', 'long_term']
for time_period in time_ranges:
    # Gets the most played songs and converts to CSV
    top_tracks = sp.current_user_top_tracks(limit=20, offset=0, time_range=time_period)
    track_ids = get_track_ids(top_tracks)
    convert_tracks_to_df(track_ids, time_period, date.today())
```

```python
def convert_artists_to_df(artist_ids, time_period, today):
    # loop over track ids 
    top_artists = []
    for i in range(len(artist_ids)):
        time.sleep(.5)
        artists = get_artist_features(artist_ids[i])
        top_artists.append(artists)
    
    # create dataset
    df = pd.DataFrame(top_artists, columns = ['artist', 'type', 'genres', 'artist_image'])
    # save to CSV
    df.to_json(f'{today}-{time_period} artists.json')
```

```python
time_ranges = ['short_term', 'medium_term', 'long_term']
for time_period in time_ranges:
    # Gets the most played artists and converts to CSV
    top_artists = sp.current_user_top_artists(limit=20, offset=0, time_range=time_period)
    artist_ids = get_artists(top_artists)
    convert_artists_to_df(artist_ids, time_period, date.today())
```
